```python
import pandas as pd
from IPython.display import Markdown
```

```python
df = pd.read_spss('ATP W41.sav')
lst = []
for name in [col for col in df.columns if col not in ['QKEY']]:
    reslt = list(df[name].unique())
    lst.append({"var": name, 'name': 'None', 'desc': reslt})
    
    
```

```python
results = '''<table>
    <th>Variable</th><th>Name</th><th>Description</th>'''
```

```python
tmp = pd.DataFrame(lst)
results = ''
def format_string(var, name, desc):
    results = f'''
    <tr>
        <td><code>{var}</code>
        <td>None</td>
        <td>
    '''
    for n, d in enumerate(desc[0]):
        results += f'<ul>{n+1} {d}</ul>'
    results += '''</td>
    </tr>'''
    
    print(results)

tmp.apply(lambda x: format_string(tmp.var, tmp.name, tmp.desc), axis=0)

print(results)

# Markdown(results)
```
